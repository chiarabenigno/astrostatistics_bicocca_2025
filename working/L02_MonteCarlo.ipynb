```python
import numpy as np
import numpy as np
from tqdm.notebook import tqdm
import pylab as plt
from matplotlib import pyplot as plt
import math
```

```python
media, sigma = 0, 2
random = np.random.normal(0, 2, 10000)
sample= np.abs(random)

def fun(x):
    return x**3

Norm = (1/2) * sigma * np.sqrt(2 * np.pi)
```

```python
montecarlo = np.mean(fun(sample))
print(montecarlo)
```

```python
integrale = Norm * montecarlo
print(integrale)
```

```python
results = []
for i in range(10, 100000, 50):
    random = np.random.normal(0, 2, i)
    sample= np.abs(random)

    def fun(x):
        return x**3

    montecarlo = np.mean(fun(sample))
    integrale = Norm * montecarlo
    results.append(integrale)
```

```python
plt.figure(figsize=(15,15))
plt.plot(range(10, 100000, 50), results)

plt.axhline(y=32, color='r', linestyle='-', linewidth=2)  # Linea orizzontale a y=2
```
